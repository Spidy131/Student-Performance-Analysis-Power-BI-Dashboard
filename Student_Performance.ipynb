```python
import pandas as pd
#import csv file
df = pd.read_csv("/content/drive/MyDrive/Student_performance5000.csv")
```

```python
#check header & data
df.info()
df.head()
```

```python
# Remove duplicate
df.drop_duplicates(inplace=True)
```

```python
#check header & data after removing duplicates
df.info()
df.head()
```

```python
#Clean Column Names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
```

```python
#check header & data
df.info()
df.head()
```

```python
#Missing name  drop (because identity is essential)
df.dropna(subset=['name'], inplace=True)
```

```python
#check header & data
df.info()
df.head()
```

```python
#Remove Outliers

#Rules:

#gpa must be between 05


for col in ['gpa']:
    df.loc[df[col] < 0, col] = 0
    df.loc[df[col] > 100, col] = 5
```

```python
#check header & data
df.info()
df.head()
```

```python
#At-Risk Student Flag
df['at_risk'] = df.apply(
    lambda row: 1 if (row['gpa'] < 2.5 or row['absences'] > 25 or row['parentalsupport'] == 0 or row['studytimeweekly'] <5 ) else 0,
    axis=1
)
```

```python
#Reorder Columns (clean final format)
df = df[[
    'studentid', 'name', 'age', 'gender',
    'degree_branch','ethnicity','parentaleducation','studytimeweekly',
    'absences','tutoring',
    'parentalsupport','extracurricular','sports','music','volunteering','gpa','gradeclass','at_risk'
]]
```

```python
#Export Cleaned Dataset
df.to_csv("cleaned_student_dataset.csv", index=False)
```
